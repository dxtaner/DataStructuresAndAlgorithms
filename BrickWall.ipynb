```python
def leastBricks(wall):
    print("=== BRICK WALL BAŞLADI ===\n")
    
    edge_count = {}
    
    for row_index, row in enumerate(wall):
        print(f"\nSatır {row_index}: {row}")
        
        position = 0
        
        # son tuğlayı dahil ETME
        for i in range(len(row) - 1):
            position += row[i]
            
            print(f"Kenar noktası: {position}")
            
            if position in edge_count:
                edge_count[position] += 1
            else:
                edge_count[position] = 1
            
            print(f"edge_count: {edge_count}")
    
    print("\n=== SON DURUM ===")
    print("Edge Count:", edge_count)
    
    if edge_count:
        max_edges = max(edge_count.values())
    else:
        max_edges = 0
    
    print(f"En fazla kesişmeyen kenar sayısı: {max_edges}")
    
    result = len(wall) - max_edges
    
    print(f"Minimum kesilen tuğla: {result}")
    
    return result
```

```python
wall = [
  [1,2,2,1],
  [3,1,2],
  [1,3,2],
  [2,4],
  [3,1,2],
  [1,3,1,1]
]

leastBricks(wall)
```
