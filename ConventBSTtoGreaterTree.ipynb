```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution:
    def convertBST(self, root):
        stack = []
        node = root
        total = 0

        while stack or node:
            # sağa git
            while node:
                stack.append(node)
                node = node.right

            node = stack.pop()

            # güncelle
            total += node.val
            node.val = total

            # sola geç
            node = node.left

        return root
```

```python
root = TreeNode(4)
root.left = TreeNode(1)
root.right = TreeNode(6)

root.left.left = TreeNode(0)
root.left.right = TreeNode(2)
root.left.right.right = TreeNode(3)

root.right.right = TreeNode(7)
```

```python
def inorder(node):
    if not node:
        return []
    return inorder(node.left) + [node.val] + inorder(node.right)


print("Önce (inorder):", inorder(root))

sol = Solution()
sol.convertBST(root)

print("Sonra (inorder):", inorder(root))
```
