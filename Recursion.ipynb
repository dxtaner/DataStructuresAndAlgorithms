```python
# Bir fonksiyonun kendi kendini çağırmasıdır.
```

```python
def count(n):
    if n == 0:
        return
    print(n)
    count(n - 1)
```

```python
count(5)
```

```python
# Factorial
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
factorial(5)
```

```python
# Sum Array
```

```python
def sum_arr(arr, depth=0):
    print("  " * depth + f"call: sum_arr({arr})")
    
    if not arr:
        print("  " * depth + "return 0")
        return 0
    
    result = arr[0] + sum_arr(arr[1:], depth + 1)
    
    print("  " * depth + f"return {result}")
    return result
```

```python
sum_arr([1, 2, 3])
```
