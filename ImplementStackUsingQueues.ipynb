```python
from collections import deque

class StackUsingQueue:
    def __init__(self):
        self.q = deque()

    def push(self, x):
        self.q.append(x)

        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self):
        if self.isEmpty():
            return "Stack boş!"
        return self.q.popleft()

    def top(self):
        if self.isEmpty():
            return "Stack boş!"
        return self.q[0]

    def isEmpty(self):
        return len(self.q) == 0

    def size(self):
        return len(self.q)
```

```python
s = StackUsingQueue()

s.push(10)
s.push(20)
s.push(30)

print(s.top())   # 30
print(s.pop())   # 30
print(s.pop())   # 20
print(s.size())  # 1
```
