Hash Table Nedir?

Hash Table:

Veriyi anahtar (key) → değer (value) şeklinde tutar

Çok hızlıdır: ortalama O(1) erişim 

Mantık:
key → hash fonksiyonu → index → değer

```python
def hashFunction(key):
    myHash = 0
    for letter in key:
        myHash = (myHash + ord(letter) * 19)
    return myHash
```

```python
class HashTable:
    
    def __init__(self, size):
        self.size = size
        self.slots = [None] * size
        self.data = [None] * size
    
    def hashFunction(self, key):
        hash_value = key % self.size
        print(f"Hash Function: {key} % {self.size} = {hash_value}")
        return hash_value
    
    def rehash(self, old_hash):
        new_hash = (old_hash + 1) % self.size
        print(f"Rehash: ({old_hash} + 1) % {self.size} = {new_hash}")
        return new_hash
    
    def put(self, key, value):
        print("\n=== PUT (EKLEME) ===")
        
        hash_value = self.hashFunction(key)
        
        if self.slots[hash_value] is None:
            print(f"Boş slot bulundu: index {hash_value}")
            self.slots[hash_value] = key
            self.data[hash_value] = value
        
        else:
            print(f"Çakışma! (Collision) index {hash_value}")
            
            if self.slots[hash_value] == key:
                print("Aynı key → değer güncelleniyor")
                self.data[hash_value] = value
            else:
                next_slot = self.rehash(hash_value)
                
                while self.slots[next_slot] is not None and self.slots[next_slot] != key:
                    print(f"Slot {next_slot} dolu → tekrar rehash")
                    next_slot = self.rehash(next_slot)
                
                print(f"Yeni slot bulundu: {next_slot}")
                self.slots[next_slot] = key
                self.data[next_slot] = value
    
    def get(self, key):
        print("\n=== GET (ARAMA) ===")
        
        start_slot = self.hashFunction(key)
        
        position = start_slot
        found = False
        stop = False
        
        while self.slots[position] is not None and not found and not stop:
            print(f"Index {position} kontrol ediliyor")
            
            if self.slots[position] == key:
                print("Key bulundu!")
                found = True
                return self.data[position]
            else:
                print("Eşleşmedi → rehash")
                position = self.rehash(position)
                
                if position == start_slot:
                    stop = True
        
        print("Key bulunamadı")
        return None
```

```python
h = HashTable(10)

h.put(23, "Taner")
h.put(33, "Ezgi")  
h.put(13, "Low")     

print(h.slots)
print(h.data)

print(h.get(23))
print(h.get(33))
print(h.get(99)) 
```
