```python
def twoSum_bruteforce(nums, target):
    print("=== BRUTE FORCE BAŞLADI ===")
    
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            print(f"Kontrol: {nums[i]} + {nums[j]}")
            
            if nums[i] + nums[j] == target:
                print(f"Bulundu: index {i}, {j}")
                return [i, j]
    
    print("Bulunamadı")
    return None
```

```python
nums = [2, 7, 11, 15]
target = 9
```

```python
twoSum_bruteforce(nums,target)
```

```python
def twoSum(nums, target):
    print("=== HASH TABLE BAŞLADI ===")
    
    hashmap = {}  # value → index
    
    for i, num in enumerate(nums):
        complement = target - num
        
        print(f"\nAdım {i}")
        print(f"Sayı: {num}")
        print(f"Aranan (complement): {target} - {num} = {complement}")
        print(f"HashMap: {hashmap}")
        
        if complement in hashmap:
            print(f"Bulundu! {complement} daha önce görülmüş")
            print(f"Indexler: {hashmap[complement]}, {i}")
            return [hashmap[complement], i]
        
        print(f"{num} hashmap'e ekleniyor")
        hashmap[num] = i
    
    print("Bulunamadı")
    return None
```

```python
twoSum(nums,target)
```
