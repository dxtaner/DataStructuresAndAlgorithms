Given an integer array nums, return true if any value appears at least twice in the array, 
and return false if every element is distinct.


Bir tamsayı dizisi verildiğinde, dizide herhangi bir değer en az iki kez geçiyorsa true değerini, tüm elemanlar birbirinden farklıysa false değerini döndür.

```python
import array as arr

myArray = arr.array("i", [1,2,3,4,5,6,7,1,9])

output = False

for i in range(len(myArray)):
    for j in range(len(myArray)):
        if i != j and myArray[i] == myArray[j]:
            output = True
            break

print(output)
```

```python
nums = [1,2,3,4,1]

for i in range(len(nums)):
    for j in range(len(nums)):
        if i != j and nums[i] == nums[j]:
            print("Duplicate bulundu:", nums[i])
```

```python
nums = [1,2,3,4,1]

seen = set()

for num in nums:
    if num in seen:
        print("Duplicate bulundu:", num)
        break
    seen.add(num)
```

```python
nums = [1,2,3,4,1]

print(len(nums) != len(set(nums)))
```
