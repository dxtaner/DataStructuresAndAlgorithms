```python
class SortingAlgorithms:
    
    def bubbleSort(self, arr):
        print("\n=== BUBBLE SORT ===")
        arr = arr.copy()
        n = len(arr)
        
        for i in range(n):
            print(f"\nPass {i+1}")
            for j in range(0, n-i-1):
                print(f"Karşılaştır: {arr[j]} - {arr[j+1]}")
                
                if arr[j] > arr[j+1]:
                    print("Swap!")
                    arr[j], arr[j+1] = arr[j+1], arr[j]
                
                print("Liste:", arr)
        
        print("Sonuç:", arr)
        return arr
    

    def selectionSort(self, arr):
        print("\n=== SELECTION SORT ===")
        arr = arr.copy()
        
        for i in range(len(arr)):
            min_index = i
            print(f"\nAdım {i}")
            
            for j in range(i+1, len(arr)):
                print(f"Karşılaştır: {arr[min_index]} - {arr[j]}")
                
                if arr[j] < arr[min_index]:
                    min_index = j
            
            print(f"Swap: {arr[i]} ↔ {arr[min_index]}")
            arr[i], arr[min_index] = arr[min_index], arr[i]
            
            print("Liste:", arr)
        
        print("Sonuç:", arr)
        return arr
    

    def insertionSort(self, arr):
        print("\n=== INSERTION SORT ===")
        arr = arr.copy()
        
        for i in range(1, len(arr)):
            key = arr[i]
            j = i - 1
            
            print(f"\nAdım {i}, key = {key}")
            
            while j >= 0 and arr[j] > key:
                print(f"{arr[j]} > {key} → sağa kaydır")
                arr[j + 1] = arr[j]
                j -= 1
            
            arr[j + 1] = key
            print("Liste:", arr)
        
        print("Sonuç:", arr)
        return arr
    

    def mergeSort(self, arr):
        print("\n=== MERGE SORT ===")
        arr = arr.copy()
        
        def merge_sort_recursive(arr):
            print(f"Split: {arr}")
            
            if len(arr) > 1:
                mid = len(arr) // 2
                left = arr[:mid]
                right = arr[mid:]
                
                merge_sort_recursive(left)
                merge_sort_recursive(right)
                
                i = j = k = 0
                
                print(f"Merge: {left} + {right}")
                
                while i < len(left) and j < len(right):
                    if left[i] < right[j]:
                        arr[k] = left[i]
                        i += 1
                    else:
                        arr[k] = right[j]
                        j += 1
                    k += 1
                
                while i < len(left):
                    arr[k] = left[i]
                    i += 1
                    k += 1
                
                while j < len(right):
                    arr[k] = right[j]
                    j += 1
                    k += 1
                
                print("Birleşti:", arr)
        
        merge_sort_recursive(arr)
        print("Sonuç:", arr)
        return arr
```

```python
sorter = SortingAlgorithms()

test_data = [5, 2, 9, 1, 6, 3]

print("\nOrijinal Liste:", test_data)

sorter.bubbleSort(test_data)
sorter.selectionSort(test_data)
sorter.insertionSort(test_data)
sorter.mergeSort(test_data)
```
