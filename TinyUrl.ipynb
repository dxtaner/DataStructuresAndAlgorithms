```python
import random
import string

class TinyURL:
    
    def __init__(self):
        self.url_map = {}   # short → long
        self.base = "http://tinyurl.com/"
        self.chars = string.ascii_letters + string.digits
    
    def generate_code(self, length=6):
        code = ''.join(random.choice(self.chars) for _ in range(length))
        print(f"Üretilen kod: {code}")
        return code
    
    def encode(self, long_url):
        print("\n=== ENCODE ===")
        print(f"Uzun URL: {long_url}")
        
        code = self.generate_code()
        
        # collision kontrolü
        while code in self.url_map:
            print("Collision! Yeni kod üretiliyor...")
            code = self.generate_code()
        
        self.url_map[code] = long_url
        
        short_url = self.base + code
        
        print(f"Kısa URL: {short_url}")
        print(f"Map durumu: {self.url_map}")
        
        return short_url
    
    def decode(self, short_url):
        print("\n=== DECODE ===")
        print(f"Kısa URL: {short_url}")
        
        code = short_url.split("/")[-1]
        print(f"Extract edilen kod: {code}")
        
        long_url = self.url_map.get(code)
        
        if long_url:
            print("Orijinal URL bulundu!")
            return long_url
        else:
            print("URL bulunamadı")
            return None
```

```python
tiny = TinyURL()

short = tiny.encode("https://example.com/very/long/url")
print(short)

original = tiny.decode(short)
print(original)
```
