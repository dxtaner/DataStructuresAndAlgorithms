Problem Tanımı

Bir stack tasarla:

push(val) → eleman ekle
pop() → en sık görülen (en yüksek frekanslı) elemanı çıkar

Eğer frekanslar eşitse:
en son eklenen (stack mantığı) çıkar

```python
class FreqStack:

    def __init__(self):
        self.freq = {}     # val → kaç kez geldi
        self.group = {}    # freq → o frekanstaki stack
        self.maxFreq = 0

        print("Başlangıç:")
        print("freq:", self.freq)
        print("group:", self.group)
        print("maxFreq:", self.maxFreq)

    def push(self, val):

        print(f"\n➕ PUSH: {val}")

        # 🔹 frekans artır
        if val in self.freq:
            self.freq[val] += 1
        else:
            self.freq[val] = 1

        f = self.freq[val]
        print(f"Frekans güncellendi: {val} → {f}")

        # 🔹 group oluştur
        if f not in self.group:
            self.group[f] = []

        # 🔹 stack'e ekle
        self.group[f].append(val)

        # 🔹 maxFreq güncelle
        if f > self.maxFreq:
            self.maxFreq = f

        # 📊 state göster
        print("freq:", self.freq)
        print("group:", self.group)
        print("maxFreq:", self.maxFreq)

    def pop(self):

        print("\n🔴 POP işlemi")

        # 🔹 en yüksek frekanstan çıkar
        val = self.group[self.maxFreq].pop()
        print(f"Çıkarılan (maxFreq={self.maxFreq}):", val)

        # 🔹 frekans azalt
        self.freq[val] -= 1

        # 🔹 grup boşsa maxFreq azalt
        if len(self.group[self.maxFreq]) == 0:
            self.maxFreq -= 1
            print("maxFreq azaldı →", self.maxFreq)

        # 📊 state göster
        print("freq:", self.freq)
        print("group:", self.group)

        return val
```

```python
fs = FreqStack()

fs.push(5)
fs.push(7)
fs.push(5)
fs.push(7)
fs.push(4)
fs.push(5)

print("\n--- POP SONUÇLARI ---")
print(fs.pop())  
print(fs.pop())  
print(fs.pop())  
print(fs.pop())  
```
