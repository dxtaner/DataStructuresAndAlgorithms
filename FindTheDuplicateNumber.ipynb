```python
def findDuplicate(nums):
    slow = nums[0]
    fast = nums[0]
    
    step = 0
    print("🔹 Phase 1: Kesişim Noktasını Bul")
    
    # 1. phase: kesişim
    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        
        print(f"Step {step}: slow={slow}, fast={fast}")
        
        if slow == fast:
            print(f"✅ Kesişim bulundu: {slow}")
            break
        
        step += 1
    
    print("\n🔹 Phase 2: Cycle başlangıcını bul (duplicate)")
    
    # 2. phase: cycle başlangıcı
    slow = nums[0]
    step = 0
    
    while slow != fast:
        slow = nums[slow]
        fast = nums[fast]
        
        print(f"Step {step}: slow={slow}, fast={fast}")
        step += 1
    
    print(f"\n🎯 Duplicate number: {slow}")
    return slow
```

```python
nums = [1,3,4,2,2]
findDuplicate(nums)
```
