```python
# 1. ListNode tanımı
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


# 2. Fonksiyonu
def getIntersectionNode(headA, headB):
    pA = headA
    pB = headB
    
    step = 0
    
    while pA != pB:
        print(f"Step {step}:")
        print(f"  pA -> {pA.val if pA else 'None'}")
        print(f"  pB -> {pB.val if pB else 'None'}")
        print("-" * 30)
        
        pA = pA.next if pA else headB
        pB = pB.next if pB else headA
        
        step += 1
    
    print(f"Step {step}:")
    print(f"  pA -> {pA.val if pA else 'None'}")
    print(f"  pB -> {pB.val if pB else 'None'}")
    print("🎯 Intersection bulundu!" if pA else "❌ Intersection yok")
    
    return pA


# 3. Test datası

# ortak node'lar
common = ListNode(7)
common.next = ListNode(8)

# A listesi
headA = ListNode(1)
headA.next = ListNode(2)
headA.next.next = ListNode(3)
headA.next.next.next = common

# B listesi
headB = ListNode(4)
headB.next = ListNode(5)
headB.next.next = common


# 4. Çalıştır
result = getIntersectionNode(headA, headB)

print("Result:", result.val if result else None)
```
