Bir array veriliyor.

Majority element:
n/2’den fazla tekrar eden elemandır.

Bu elemanı bul.

```python
def majority_element(nums):
    for num in nums:
        if nums.count(num) > len(nums)//2:
            return num
```

```python

myList = [3,3,4,2,3,3,5]
```

```python
majority_element(myList)
```
