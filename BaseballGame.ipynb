```python
def calPoints(ops):
    stack = []

    for op in ops:
        
        if op == "C":
            stack.pop()  
            # Son skoru sil

        elif op == "D":
            stack.append(2 * stack[-1])  
            # Son skorun 2 katını ekle

        elif op == "+":
            stack.append(stack[-1] + stack[-2])  
            # Son iki skorun toplamını ekle

        else:
            stack.append(int(op))  
            # Sayıyı ekle

    return sum(stack)


```

```python
# Test
ops = ["5","2","C","D","+"]
print(calPoints(ops))   # 30
```

```python
ops = ["2","3","D","D","+"]
print(calPoints(ops))   # 41
```
