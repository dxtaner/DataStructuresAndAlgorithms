```python
class Node():
    def __init__(self,value):
        self.value = value
        self.nextNode = None
```

```python
firstNode = Node(10)
```

```python
secondNode = Node(20)
```

```python
thirdNode = Node(30)
```

```python
firstNode.nextNode = secondNode
```

```python
secondNode.nextNode = thirdNode
```

```python
firstNode.nextNode.nextNode.value
```
