```python
class Deque:
    def __init__(self):
        self.items = []

    def addRight(self, element):
        self.items.append(element)

    def addLeft(self, element):
        self.items.insert(0, element)

    def removeRight(self):
        if self.isEmpty():
            return "Deque boş!"
        return self.items.pop()

    def removeLeft(self):
        if self.isEmpty():
            return "Deque boş!"
        return self.items.pop(0)

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
d = Deque()

d.addRight(10)
d.addRight(20)
d.addLeft(5)
```

```python
print("Deque:", d.items)        

print("removeRight:", d.removeRight()) 
print("removeLeft:", d.removeLeft())    

print("Son durum:", d.items)   
print("Boyut:", d.size())      
print("Boş mu:", d.isEmpty())   
```
