```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def removeNthFromEnd(head, n):
    dummy = ListNode(0)
    dummy.next = head
    
    fast = dummy
    slow = dummy
    
    # 1. fast pointer'ı n+1 adım ilerlet
    for _ in range(n + 1):
        fast = fast.next
    
    # 2. fast sona gidene kadar birlikte ilerle
    while fast:
        fast = fast.next
        slow = slow.next
    
    # 3. silme işlemi
    slow.next = slow.next.next
    
    return dummy.next
```

```python
def build_list(arr):
    dummy = ListNode(0)
    current = dummy
    for x in arr:
        current.next = ListNode(x)
        current = current.next
    return dummy.next

def print_list(head):
    while head:
        print(head.val, end=" -> ")
        head = head.next
    print("None")
```

```python
def debug_print(head, label):
    print(label, end=": ")
    print_list(head)

head = build_list([1,2,3,4,5])
debug_print(head, "Before")

head = removeNthFromEnd(head, 2)

debug_print(head, "After")
```
