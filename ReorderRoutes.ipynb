Problem Tanımı

Elimizde n adet şehir bulunmaktadır (0 → n-1).

Bize verilen connections listesi, şehirler arasındaki yönlü yolları gösterir.

Amaç

Tüm şehirlerden şehir 0’a ulaşılabilir hale getirmek için
minimum kaç yolu ters çevirmemiz gerektiğini bulmak

```python
class Solution:
    def minReorder(self, n, connections):
        print("Graph oluşturuluyor...")

        # Graph oluştur
        graph = {i: [] for i in range(n)}

        # Edge'leri ekle
        for a, b in connections:
            print(f" Edge ekleniyor: {a} -> {b} (cost=1), {b} -> {a} (cost=0)")
            graph[a].append((b, 1))
            graph[b].append((a, 0))

        print("\n Graph:")
        for k, v in graph.items():
            print(f"{k} -> {v}")

        visited = set()
        self.changes = 0

        def dfs(node):
            print(f"\n Node ziyaret edildi: {node}")
            visited.add(node)

            for neighbor, cost in graph[node]:
                print(f"   {node} → {neighbor} (cost={cost})")

                if neighbor not in visited:
                    if cost == 1:
                        print(f"    Yön ters → değiştirilecek (+1)")
                    else:
                        print(f"   Zaten doğru yön")

                    self.changes += cost
                    dfs(neighbor)
                else:
                    print(f"    Zaten ziyaret edilmiş")

        print("\n DFS başlıyor (node 0)")
        dfs(0)

        print(f"\n Toplam değişiklik sayısı: {self.changes}")
        return self.changes
```

```python
n = 6
connections = [[0,1],[1,3],[2,3],[4,0],[4,5]]

sol = Solution()
print(sol.minReorder(n, connections)) 
```
