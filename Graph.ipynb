```python
class Graph:
    def __init__(self):
        self.adjDict = {}

    def addVertex(self, vertex):
        print(f"\n[addVertex] {vertex} ekleniyor...")
        if vertex not in self.adjDict:
            self.adjDict[vertex] = []
            print(f" {vertex} eklendi")
            return True
        print(f" {vertex} zaten var")
        return False

    def addEdge(self, v1, v2):
        print(f"\n[addEdge] {v1} <-> {v2}")
        if v1 in self.adjDict and v2 in self.adjDict:
            if v2 not in self.adjDict[v1]:
                self.adjDict[v1].append(v2)
            if v1 not in self.adjDict[v2]:
                self.adjDict[v2].append(v1)

            print(f" Edge eklendi: {v1} - {v2}")
            return True
        print(" Vertex bulunamadı")
        return False

    def removeEdge(self, v1, v2):
        print(f"\n[removeEdge] {v1} - {v2} siliniyor...")
        if v1 in self.adjDict and v2 in self.adjDict:
            if v2 in self.adjDict[v1]:
                self.adjDict[v1].remove(v2)
            if v1 in self.adjDict[v2]:
                self.adjDict[v2].remove(v1)

            print(f" Edge silindi: {v1} - {v2}")
            return True
        print(" Vertex bulunamadı")
        return False

    def removeVertex(self, vertex):
        print(f"\n[removeVertex] {vertex} siliniyor...")
        if vertex in self.adjDict:
            for neighbor in list(self.adjDict[vertex]):
                print(f"   -> {neighbor} bağlantısı siliniyor")
                self.adjDict[neighbor].remove(vertex)

            del self.adjDict[vertex]
            print(f" {vertex} tamamen silindi")
            return True

        print(" Vertex bulunamadı")
        return False

    def printGraph(self):
        print("\n Graph Durumu:")
        for vertex in self.adjDict:
            print(vertex, "->", self.adjDict[vertex])
```

```python
g = Graph()
```

```python
g.addVertex("Ankara")
g.addVertex("Konya")
g.addVertex("Bursa")
g.addVertex("Afyon")

g.printGraph()
```

```python
g.addEdge("Ankara", "Konya")
g.addEdge("Konya", "Afyon")
g.addEdge("Konya", "Ankara")
g.addEdge("Afyon", "Konya")
g.addEdge("Afyon", "Bursa")
g.addEdge("Bursa", "Afyon")

g.printGraph()
```

```python
g.removeVertex("Afyon")

g.printGraph()
```
