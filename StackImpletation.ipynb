```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, element):
        self.items.append(element)

    def pop(self):
        if self.isEmpty():
            return "Stack boş!"
        return self.items.pop()

    def showLast(self):
        if self.isEmpty():
            return "Stack boş!"
        return self.items[-1]

    def size(self):
        return len(self.items)

    def isEmpty(self):
        return len(self.items) == 0



```

```python
s = Stack()

s.push(10)
s.push(20)
s.push(30)
```

```python
print("Son eleman:", s.showLast())  
print("Pop:", s.pop())              
print("Boyut:", s.size())           
print("Boş mu:", s.isEmpty())      
```
