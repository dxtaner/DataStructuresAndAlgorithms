```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    result = [0] * n
    stack = []  

    for i in range(n):
        print(f"\nGün {i} | Sıcaklık: {temperatures[i]}")
        
        # Stack kontrol
        while stack and temperatures[i] > temperatures[stack[-1]]:
            prev_index = stack.pop()
            result[prev_index] = i - prev_index

            print(f"  ➤ {temperatures[prev_index]} < {temperatures[i]}")
            print(f"  ➤ Gün {prev_index} için cevap bulundu: {result[prev_index]} gün sonra")

        stack.append(i)
        print(f"  ➤ Stack durumu (index): {stack}")
        print(f"  ➤ Stack sıcaklıkları: {[temperatures[idx] for idx in stack]}")
        print(f"  ➤ Result: {result}")

    return result


# Test
temps = [73,74,75,71,69,72,76,73]
sonuc = dailyTemperatures(temps)

print("\nFinal Sonuç:", sonuc)
```
