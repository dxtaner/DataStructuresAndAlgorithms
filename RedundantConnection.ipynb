Problem Tanımı

Elimizde n adet node’dan oluşan bir graph var.
Başlangıçta bu graph bir tree (ağaç) yapısındadır:

Bağlantılıdır (connected)
Cycle (döngü) yoktur

Ancak daha sonra fazladan bir edge ekleniyor.

Amaç

Graph’ı tekrar bir tree haline getirmek için
hangi edge’i kaldırmamız gerektiğini bulmak

```python
class Solution:
    def findRedundantConnection(self, edges):
        parent = {}

        # find (path compression ile)
        def find(x):
            if parent[x] != x:
                print(f"find({x}) → parent[{x}]={parent[x]} → yukarı çıkılıyor")
                parent[x] = find(parent[x])
            return parent[x]

        # union işlemi
        def union(x, y):
            rootX = find(x)
            rootY = find(y)

            print(f"\n Union({x}, {y})")
            print(f"   rootX = {rootX}, rootY = {rootY}")

            # Eğer aynı root ise cycle var
            if rootX == rootY:
                print(f" Cycle detected! ({x},{y}) redundant edge")
                return False

            # union işlemi
            parent[rootY] = rootX
            print(f" {rootY} artık {rootX}'e bağlandı")
            return True

        # parent initialize
        for u, v in edges:
            if u not in parent:
                parent[u] = u
            if v not in parent:
                parent[v] = v

        print("Başlangıç parent:", parent)

        # edge'leri işle
        for u, v in edges:
            if not union(u, v):
                print(f"\n SONUÇ: Redundant Edge = [{u}, {v}]")
                return [u, v]
```

```python
edges = [[1,2],[2,3],[3,4],[1,4],[1,5]]
```

```python
sol = Solution()
print(sol.findRedundantConnection(edges))
```
