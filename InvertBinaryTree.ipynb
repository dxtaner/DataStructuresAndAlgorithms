```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(7)

root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right.left = TreeNode(6)
root.right.right = TreeNode(9)
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
```

```python
def invert_debug(root, depth=0):
    if not root:
        return None
    
    print("  " * depth + f"Node: {root.val} (swap)")
    
    root.left, root.right = root.right, root.left
    
    invert_debug(root.left, depth+1)
    invert_debug(root.right, depth+1)
    
    return root
```

```python
print("Before:")
inorder(root)

print("\n\nInvert:")
invert_debug(root)

print("\nAfter:")
inorder(root)
```
