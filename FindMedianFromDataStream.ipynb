Veri akışı (stream) şeklinde sayılar geliyor.
Bizden istenen:

addNum(num) → yeni sayı ekle
findMedian() → o ana kadar gelen sayıların median’ını bul

```python
class MedianFinder:

    def __init__(self):
        self.nums = []
        print("Başlangıç:", self.nums)

    def addNum(self, num):

        print(f"\n➕ Ekleniyor: {num}")
        print("Önce liste:", self.nums)

        # Binary search ile doğru index bul
        left, right = 0, len(self.nums)

        step = 0
        while left < right:
            mid = (left + right) // 2
            print(f"Adım {step} | left={left}, right={right}, mid={mid}")

            if self.nums[mid] < num:
                print(f"    {self.nums[mid]} < {num} → sağa git")
                left = mid + 1
            else:
                print(f"    {self.nums[mid]} ≥ {num} → sola git")
                right = mid

            step += 1

        print(f"Insert index: {left}")

        # ekleme
        self.nums.insert(left, num)

        print("Son liste:", self.nums)

    def findMedian(self):

        n = len(self.nums)
        mid = n // 2

        print("\n📊 Median hesaplanıyor...")
        print("Liste:", self.nums)

        if n % 2 == 1:
            print(f"Tek sayıda → median = {self.nums[mid]}")
            return self.nums[mid]

        median = (self.nums[mid - 1] + self.nums[mid]) / 2
        print(f"Çift sayıda → ({self.nums[mid-1]} + {self.nums[mid]}) / 2 = {median}")
        return median
```

```python
mf = MedianFinder()

mf.addNum(1)
print(mf.findMedian())   

mf.addNum(2)
print(mf.findMedian())   

mf.addNum(3)
print(mf.findMedian())   

mf.addNum(4)
print(mf.findMedian())  
```
