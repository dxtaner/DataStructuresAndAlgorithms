```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class BST:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if not self.root:
            self.root = Node(value)
            return

        cur = self.root
        while True:
            if value < cur.value:
                if not cur.left:
                    cur.left = Node(value)
                    return
                cur = cur.left
            else:
                if not cur.right:
                    cur.right = Node(value)
                    return
                cur = cur.right

    def contains(self, value):
        cur = self.root
        while cur:
            if value == cur.value:
                return True
            cur = cur.left if value < cur.value else cur.right
        return False
```

```python
bst = BST()

for x in [10, 5, 15]:
    bst.insert(x)

print(bst.contains(5))   # True
print(bst.contains(99))  # False
```
