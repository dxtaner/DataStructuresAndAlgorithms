Problem Tanımı

Verilen 2D noktalar arasından:

Orijine (0,0) en yakın K tane noktayı bul

```python
import heapq

class KClosestPoints:

    def kClosest_heap(self, points, k):
        # Heap'e (distance, point) olarak ekle
        heap = []
        
        for (x, y) in points:
            dist = x*x + y*y
            heap.append((dist, [x, y]))
        
        # Heap yap
        heapq.heapify(heap)
        
        result = []
        
        # En küçük k elemanı al
        for _ in range(k):
            result.append(heapq.heappop(heap)[1])
        
        return result
```

```python
points = [[3,3],[5,-1],[-2,4]]
k = 2

solver = KClosestPoints()

print("Heap:", solver.kClosest_heap(points, k))
```
