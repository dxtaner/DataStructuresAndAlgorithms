```python
import sys
n = 30
myDynamicArray = []

for i in range(0,n):
    myLenght = len(myDynamicArray)
    myByte = sys.getsizeof(myDynamicArray)
    print(f"Lenght: {myLenght} Byte: {myByte}")
    myDynamicArray.append(n)
```
