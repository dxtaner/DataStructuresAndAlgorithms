```python
def singleNumber():
    myList = [1,5,7,4,7]
    result = 0
    
    print("Başlangıç result =", result)
    print("-" * 30)

    for num in myList:
        print(f"İşlem: {result} ^ {num}")
        result = result ^ num
        print(f"Yeni result = {result}")
        print("-" * 30)
    
    return result

print("Sonuç:", singleNumber())
```
