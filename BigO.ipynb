# Big-O Notation, Time Complexity & Space Complexity

Bu notebook, algoritmaların performansını analiz etmek için kullanılan Big-O Notation kavramını açıklar.


## 🔹 Big-O Notation Nedir?

Big-O Notation, bir algoritmanın giriş boyutu (n) büyüdükçe performansının nasıl değiştiğini gösterir.

Genellikle:
- Worst-case (en kötü durum) analiz edilir
- Zaman ve alan karmaşıklığı ölçülür


## ⏱ Time Complexity (Zaman Karmaşıklığı)

Algoritmanın çalışma süresini ifade eder.

### Yaygın Notasyonlar:

- O(1) → Sabit zaman
- O(log n) → Logaritmik
- O(n) → Lineer
- O(n log n) → Verimli sıralama
- O(n²) → Nested loop
- O(2ⁿ) → Üstel (çok yavaş)


Kod Örnekleri

O(1)

```python
def get_first(arr):
    return arr[0]

print(get_first([1,2,3,4]))
```

O(n)

```python
def print_all(arr):
    for x in arr:
        print(x)

print_all([1,2,3,4])
```

O(n²)

```python
def print_pairs(arr):
    for i in arr:
        for j in arr:
            print(i, j)

print_pairs([1,2,3])
```

O(log n) (Binary Search)

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return -1

print(binary_search([1,2,3,4,5,6], 4))
```

## 💾 Space Complexity (Alan Karmaşıklığı)

Algoritmanın ne kadar bellek kullandığını gösterir.


O(1)

```python
def sum_two(a, b):
    return a + b

print(sum_two(5, 10))
```

O(n)

```python
def copy_array(arr):
    new_arr = []
    for x in arr:
        new_arr.append(x)
    return new_arr

print(copy_array([1,2,3]))
```

Recursive Space (Stack)

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

## 📏 Big-O Kuralları

1. Sabitleri yok say:
   O(2n) → O(n)

2. En büyük terimi al:
   O(n² + n) → O(n²)

3. İç içe döngüler çarpılır:
   O(n * n) → O(n²)

4. Ayrı döngüler toplanır:
   O(n + n) → O(n)


```python
# Duplicate kontrolü

def has_duplicate(arr):
    return len(arr) != len(set(arr))

print(has_duplicate([1,2,3,4]))
print(has_duplicate([1,2,2,3]))
```
